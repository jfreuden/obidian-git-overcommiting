# Obsidian Git Commit Analysis
I noticed that the more intensively I am using my Obsidian vault, the more commits the `obsidian-git` plugin makes, which is polluting my contribution graph.

If I can collapse the commits down to reduce the pollution, I can get a happy medium between granularly tracking my modifications and reducing the noise in my contribution graph.

# Goal
Find out how many commits I'll knock off if I squash the commits together, dependent on what size windows I want to use.

## Windowing
A naive binning would be to just slice all the commits into regular "hour" buckets. However, I really would like to clump commits into each other instead, such that no two commits have less than an hour of time between them.

### Example:
- Commit A at 10:30
- Commit B at 11:11
- Commit C at 12:08


In the Naive binning example, commits A, B and C will have separate bins. In the clumped binning example, they will be merged into the same bin, centered around commit B.

Or could do it so that all commits must be within an hour of each other, in which case the easiest way would be a rolling window, and then it's grab a commit
